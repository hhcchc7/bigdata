```
install.packages("ggplot2")
library(SportsAnalytics)
library(ggplot2)

NBA1516 <- fetch_NBAPlayerStatistics("15-16")
```

```
qplot(TotalMinutesPlayed, GamesPlayed, data = NBA1516,color=Position)
```

```
qplot(TotalMinutesPlayed, GamesPlayed, data = NBA1516,
geom = c("point", "smooth"))
#x=TotalMinutesPlayed y=GamesPlayed
```

```
#畫GamesPlayed的直方圖/ fill = Position 並用守備位置Position著色
qplot(GamesPlayed, data = NBA1516, fill = Position)
#x=GamesPlayed
```

```
#facets = .~ Position 用守備位置Position分群畫圖(直向)
qplot(TotalMinutesPlayed, GamesPlayed, data = NBA1516, facets = Position ~ .)
#x=TotalMinutesPlayed
```

```
#facets = .~ Position 用守備位置Position分群畫圖(橫向)
qplot(TotalMinutesPlayed, GamesPlayed, data = NBA1516, facets = . ~ Position)
#x=TotalMinutesPlayed
```

```
#facets = .~ Position 用守備位置Position分群畫圖(直向)
#binwith = 10 每10分一組
qplot(GamesPlayed, data = NBA1516, facets = Position ~ . , binwidth = 10)
#x=GamesPlayed
```

```
#facets = .~ Position 用守備位置Position分群畫圖(直向)
#binwith = 100 每100分一組
qplot(GamesPlayed, data = NBA1516, facets = Position ~ . , binwidth = 100)
#x=GamesPlayed
```
